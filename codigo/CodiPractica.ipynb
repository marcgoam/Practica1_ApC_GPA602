```python
#Importem les llibreries necesaries i carreguem el dataset

from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

# Visualitzarem només 3 decimals per mostra
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Funcio per a llegir dades en format csv
def load_dataset(path):
    dataset = pd.read_csv(path, header=0, delimiter=',')
    return dataset

# Carreguem dataset
dataset = load_dataset('indicator-life_expectancy_at_birth.csv').dropna().reset_index(drop=True)

data = dataset.values

x = data
y = data[:, 2016-1800+1]
```

```python
#Funcions per aplicar la refressió
def mse(v1, v2):
    return ((v1 - v2)**2).mean()

def regression(x, y):
    # Creem un objecte de regressió de sklearn
    regr = LinearRegression()

    # Entrenem el model per a predir y a partir de x
    regr.fit(x, y)

    # Retornem el model entrenat
    return regr

def standarize(x_train):
    x_train = x_train.astype(int)
    mean = x_train.mean(0)
    std = x_train.std(0)
    x_t = x_train - mean[None, :]
    x_t /= std[None, :]
    return x_t
```

```python
from sklearn.metrics import r2_score
from os import remove

# Extraiem el primer atribut de x i canviem la mida a #exemples, #dimensions de l'atribut.
# En el vostre cas, haureu de triar un atribut com a y, i utilitzar la resta com a x.


def primeraRegresio():
    
    try:   
        remove("./results/errorsQuadratics.txt")
    except:
        pass
    
    x_train, x_test, y_train, y_test = train_test_split(x[:,1:],y,train_size=0.8,random_state = 1234,shuffle=True)
    prediccio = []
    entreno = []

    for i in range(x_train.shape[1]):
        entreno.append(regression(x_train[:,i].reshape(-1,1), y_train))
    entreno = np.array(entreno)
    for i in range(x_test.shape[1]):
        prediccio.append(entreno[i].predict(x_test[:,i].reshape(-1,1)))
    
    for i in range(x_test.shape[1]):
        ax = plt.scatter(x_test[:,i], y_test)
        plt.plot(x_test[:,i], prediccio[i], 'r')
        plt.savefig('./img/regresions/noNormalitzat/' + str(1800+i) + '.png')
        plt.close()
        
    for i, predicted in enumerate(prediccio):
        MSE = mse(predicted, y_test)
        r2 = r2_score(predicted, y_test)
        with open("./results/errorsQuadratics.txt",'a') as s:
            s.write("Atribut " + str(i+1) + " Any: " + str(1800+i) + " :\n")
            s.write("   MSE: "+ str(MSE) + "   R2 score: " + str(r2) + "\n")
        


    
primeraRegresio()


```

```python
#Regresió normalitzada


def segonaRegresio():
    
    try:
        remove("./results/errorsQuadraticsNormalitzat.txt")
    except:
        pass
    
    #Normalitzem les dades
    data_aux = data[:,1:]
    data_stand = standarize(data_aux)
    x_norm = data_stand
    y_norm = data_stand[:, 2016-1800]

    x_train, x_test, y_train, y_test = train_test_split(x_norm[:,1:],y_norm,train_size=0.8,random_state = 1234,shuffle=True)
    
    prediccio = []
    entreno = []

    for i in range(x_train.shape[1]):
        entreno.append(regression(x_train[:,i].reshape(-1,1), y_train))
    entreno = np.array(entreno)
    for i in range(x_test.shape[1]):
        prediccio.append(entreno[i].predict(x_test[:,i].reshape(-1,1)))
    
    for i in range(x_test.shape[1]):
        ax = plt.scatter(x_test[:,i], y_test)
        plt.plot(x_test[:,i], prediccio[i], 'r')
        plt.savefig('./img/regresions/normalitzat/' + str(1800+i) + '.png')
        plt.close()
        
    for i, predicted in enumerate(prediccio):
        MSE = mse(predicted, y_test)
        r2 = r2_score(predicted, y_test)
        with open("./results/errorsQuadraticsNormalitzat.txt",'a') as s:
            s.write("Atribut " + str(i+1) + " Any: " + str(1800+i) + " :\n")
            s.write("   MSE: "+ str(MSE) + "   R2 score: " + str(r2) + "\n")
    

segonaRegresio()
```
